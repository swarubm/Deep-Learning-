9. design and implement a Recurrent neural network (RNN) for generating text
sequences “assume text sequence rhyme twinkle twinkle little star”

```
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
# --- DATA ---
text = """
twinkle twinkle little star
how i wonder what you are
up above the world so high
like a diamond in the sky
twinkle twinkle little star
how i wonder what you are
"""

text = text.lower().split() # word-level
# Build vocabulary
words = sorted(set(text))
word2idx = {w:i for i,w in enumerate(words)}
idx2word = {i:w for w,i in word2idx.items()}
data = [word2idx[w] for w in text]
seq_len = 5 # number of previous words to consider
X = np.array([data[i:i+seq_len] for i in range(len(data)-seq_len)])
y = np.array([data[i+seq_len] for i in range(len(data)-seq_len)])
# --- MODEL ---
model = Sequential([
  Embedding(len(words), 50),
  LSTM(128),
  Dense(len(words), activation='softmax')

])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
model.fit(X, y, epochs=300, verbose=0)
# --- GENERATE ---
def generate(start="twinkle twinkle little star", length=30, temperature=0.7):
  result = start.lower().split()
  for _ in range(length):
    x = np.array([[word2idx.get(w, 0) for w in result[-seq_len:]]])
    preds = model(x)[0].numpy() / temperature
    next_id = tf.random.categorical(tf.math.log(preds[None, :]),
                                    1).numpy()[0][0]
    result.append(idx2word[next_id])
  return " ".join(result)
# --- OUTPUT ---
print("\nGenerated poem:\n")
print(generate("twinkle twinkle little star", 40, temperature=0.6))
```
